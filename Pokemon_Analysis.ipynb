```python
df = pd.read_csv('Pokemon.csv') #To read pokemon data from CSV file

# Calculate total stats for each Pokemon
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

# Loop through each generation
for gen in df['Generation'].unique():
    df_gen = df[df['Generation'] == gen] # Filter data for current generation
    df_sorted = df_gen.sort_values('Total', ascending=False) # Sort Pokemon by total stats

    # Get strongest and weakest Pokemon
    strongest = df_sorted.iloc[0]
    weakest = df_sorted.iloc[-1]

    # Print results
    print(f"Generation {gen}:")
    print(f"  Strongest Pokemon: {strongest['Name']} (Total: {strongest['Total']})")
    print(f"  Weakest Pokemon: {weakest['Name']} (Total: {weakest['Total']})\n")
```

```python
# To discover pokemon with the strongest HP
strongest_hp_pokemon = df.loc[df['HP'].idxmax()]
print("\nPokemon with the strongest HP:")
print(f"Name: {strongest_hp_pokemon['Name']}, HP: {strongest_hp_pokemon['HP']}")

# To discover pokemon with the strongest Attack
strongest_attack_pokemon = df.loc[df['Attack'].idxmax()]
print("\nPokemon with the strongest Attack:")
print(f"Name: {strongest_attack_pokemon['Name']}, Attack: {strongest_attack_pokemon['Attack']}")

# To discover pokemon with the strongest Defense
strongest_defense_pokemon = df.loc[df['Defense'].idxmax()]
print("\nPokemon with the strongest Defense:")
print(f"Name: {strongest_defense_pokemon['Name']}, Defense: {strongest_defense_pokemon['Defense']}")

# To discover pokemon with the strongest Speed
strongest_speed_pokemon = df.loc[df['Speed'].idxmax()]
print("\nPokemon with the strongest Speed:")
print(f"Name: {strongest_speed_pokemon['Name']}, Speed: {strongest_speed_pokemon['Speed']}")
```

```python
# To Discover Pokemon with the weakest HP
weakest_hp_pokemon = df.loc[df['HP'].idxmin()]
print("\nPokemon with the weakest HP:")
print(f"Nama: {weakest_hp_pokemon['Name']}, HP: {weakest_hp_pokemon['HP']}")

# To Discover Pokemon with the weakest Attack
weakest_attack_pokemon = df.loc[df['Attack'].idxmin()]
print("\nPokemon with the weakest Attack")
print(f"Nama: {weakest_attack_pokemon['Name']}, Attack: {weakest_attack_pokemon['Attack']}")

# To Discover Pokemon with the weakest Defense
weakest_defense_pokemon = df.loc[df['Defense'].idxmin()]
print("\nPokemon with the weakest Defense:")
print(f"Nama: {weakest_defense_pokemon['Name']}, Defense: {weakest_defense_pokemon['Defense']}")

# To Discover Pokemon with the weakest Speed
weakest_speed_pokemon = df.loc[df['Speed'].idxmin()]
print("\nPokemon with the weakest Speed:")
print(f"Nama: {weakest_speed_pokemon['Name']}, Speed: {weakest_speed_pokemon['Speed']}")
```

#Visualization

```python
# Read Pokemon data from CSV file
df = pd.read_csv('Pokemon.csv')

# Count the number of legendary and non-legendary Pokemon
legendary_count = df['Legendary'].sum()
non_legendary_count = len(df) - legendary_count

# Data for the pie chart
labels = ['Legendary', 'Non-Legendary']
sizes = [legendary_count, non_legendary_count]

# Create the pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax.set_title('Comparison of Legendary and Non-Legendary Pokemon')

plt.show()
```

```python
# Read Pokemon data from the CSV file, make sure 'Pokemon.csv' is in the same directory
df = pd.read_csv('Pokemon.csv')

# Count Pokemon with only one type (Type 2 is missing)
single_type_count = df['Type 2'].isna().sum()

# Count Pokemon with two types (Type 2 is not missing)
dual_type_count = df['Type 2'].notna().sum()

# Data for the pie chart
labels = ['Single Type', 'Dual Type']
sizes = [single_type_count, dual_type_count]

# Create the pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Make the pie chart a circle
ax.set_title('Proportion of Pokemon with Single vs. Dual Types')

plt.show()

print("Number of Pokemon with single type:", single_type_count)
print("Number of Pokemon with dual types:", dual_type_count)
```

```python
# Read Pokemon data from the CSV file, make sure 'Pokemon.csv' is in the same directory
df = pd.read_csv('Pokemon.csv')

# Combine Type 1 and Type 2 columns, removing missing values
all_types = pd.concat([df['Type 1'], df['Type 2']]).dropna()

# Count the number of each type
type_counts = all_types.value_counts()

# Create a bar chart
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
type_counts.plot(kind='bar')
plt.xlabel('Pokemon Type')
plt.ylabel('Number of Pokemon')
plt.title('Number of Pokemon by Type')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data Pokemon dari file CSV
df = pd.read_csv('Pokemon.csv')

# Hitung total statistik untuk setiap Pokemon
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

# Urutkan Pokemon berdasarkan total statistik
df_sorted = df.sort_values('Total', ascending=False)

# Ambil Pokemon terkuat dan terlemah
strongest = df_sorted.iloc[0]
weakest = df_sorted.iloc[-1]

# Buat grafik batang
fig, ax = plt.subplots(figsize=(10, 6))  # Sesuaikan ukuran grafik jika diperlukan

# Statistik yang akan ditampilkan
stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
x = range(len(stats))  # Posisi batang pada sumbu x

# Buat batang untuk Pokemon terkuat
ax.bar(x, [strongest[stat] for stat in stats], width=0.4, label=strongest['Name'], align='center')

# Buat batang untuk Pokemon terlemah
ax.bar([i + 0.4 for i in x], [weakest[stat] for stat in stats], width=0.4, label=weakest['Name'], align='center')

# Atur label sumbu dan judul
ax.set_xticks(x)
ax.set_xticklabels(stats)
ax.set_ylabel('Statistical Value')
ax.set_title('Statistical Comparison of the Strongest and Weakest Pokemon')
ax.legend()

plt.show()
```
